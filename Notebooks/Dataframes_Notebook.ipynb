```python
#Import the necessary libraries: pandas
import pandas as pd
```

```python
#load the CSV file
df_ventas = pd.read_csv("C:/Users/juane/OneDrive/Escritorio/Proyectos Python/Project Bakery/Data/Ventas.csv")
df_calendario = pd.read_csv("C:/Users/juane/OneDrive/Escritorio/Proyectos Python/Project Bakery/Data/Calendario.csv")
```

```python
# Convert date columns to datetime
df_ventas['FechaVenta'] = pd.to_datetime(df_ventas['FechaVenta'], format='mixed', dayfirst=True)
df_calendario['Fecha'] = pd.to_datetime(df_calendario['Fecha'], format='mixed', dayfirst=True)

# Join the DataFrames
df_unido = pd.merge(df_ventas, df_calendario, left_on='FechaVenta', right_on='Fecha', how='left')

# Fill null values ​​in 'Holiday' column with 'Non-holiday'
df_unido['Festivo'].fillna('Día no festivo', inplace=True)

# Create the new columns 'Year' and 'Month_Number'
df_unido['Año'] = df_unido['FechaVenta'].dt.year
df_unido['Numero_Mes'] = df_unido['FechaVenta'].dt.month

# Add the 'Day_Week' column
df_unido['Dia_Semana'] = df_unido['FechaVenta'].dt.day_name()

# Add the 'It's_Festive' column
df_unido['Es_Festivo'] = df_unido['Festivo'].apply(lambda x: 'Sí' if x != 'Día no festivo' else 'No')

# Get the list of columns
cols = df_unido.columns.tolist()

# Find the position of the 'SalesDate' column
pos_fecha = cols.index('FechaVenta')

# Insert the new columns 'Year' and 'Month_Number' after 'SalesDate'
cols.insert(pos_fecha + 1, cols.pop(cols.index('Año')))
cols.insert(pos_fecha + 2, cols.pop(cols.index('Numero_Mes')))

# Find the position of the 'Holiday' column
pos_festivo = cols.index('Festivo')

# Insert new column 'It's_Holiday' after 'Holiday'
cols.insert(pos_festivo + 1, cols.pop(cols.index('Es_Festivo')))

# Reorder the DataFrame
df_unido = df_unido[cols]
```

```python
# Initialize empty lists for each column
id_provincia = []
id_estacion = []
fecha = []
año = []
dia = []
temp_media = []
humedad_media = []
vel_viento = []
precipitacion = []

# Read the file line by line
with open("C:\\Users\\juane\\OneDrive\\Escritorio\\Datos\\Datos_csv\\clima.csv", encoding='iso-8859-1') as file:
    next(file)  # Saltar la primera línea (encabezados)
    for line in file:
        # Dividir la línea por el delimitador ';'
        parts = line.strip().split(';')
        if len(parts) == 9:  # Asegurarse de que la línea tenga el número correcto de columnas
            id_provincia.append(parts[0])
            id_estacion.append(parts[1])
            fecha.append(parts[2])
            año.append(parts[3])
            dia.append(parts[4])
            temp_media.append(parts[5])
            humedad_media.append(parts[6])
            vel_viento.append(parts[7])
            precipitacion.append(parts[8])

# Create the DataFrame manually
df_clima = pd.DataFrame({
    'IdProvincia': id_provincia,
    'IdEstacion': id_estacion,
    'Fecha': fecha,
    'Año': año,
    'Dia': dia,
    'Temp Media (ºC)': temp_media,
    'Humedad Media (%)': humedad_media,
    'Velviento (m/s)': vel_viento,
    'Precipitación (mm)': precipitacion
})
```

```python
# Show the first rows of the DataFrame
df_clima.head()
```

```python
# Convert date columns to datetime
df_unido['Fecha'] = pd.to_datetime(df_unido['Fecha'], dayfirst=True)
df_clima['Fecha'] = pd.to_datetime(df_clima['Fecha'], dayfirst=True)

# Join the DataFrames in the 'Date' column
df = pd.merge(df_unido, df_clima[['Fecha', 'Temp Media (ºC)', 'Humedad Media (%)', 'Velviento (m/s)', 'Precipitación (mm)']], on='Fecha', how='left')
```

```python
df.head()
```

```python
#Delete 'Date' column
df.drop(columns=['Fecha'], inplace=True)

# Assign new names to columns
df.columns = ['Familia', 'Tipo', 'Fecha', 'Año', 'Mes', 'Hora', 'Articulo', 'Cantidad', 'Precio', 'Importe', 'Festivo', 'Es_Festivo', 'Dia_semana','Temperatura','Humedad','Viento','Precipitación']
```

```python
# Make sure the columns are of type string
df['Temperatura'] = df['Temperatura'].astype(str)
df['Humedad'] = df['Humedad'].astype(str)
df['Viento'] = df['Viento'].astype(str)
df['Precipitación'] = df['Precipitación'].astype(str)

# Replace unwanted characters and convert to float
df['Temperatura'] = df['Temperatura'].str.replace(r'[^\d.,]', '', regex=True).str.replace(',', '.').replace('', '0').astype(float)
df['Humedad'] = df['Humedad'].str.replace(r'[^\d.,]', '', regex=True).str.replace(',', '.').replace('', '0').astype(float)
df['Viento'] = df['Viento'].str.replace(r'[^\d.,]', '', regex=True).str.replace(',', '.').replace('', '0').astype(float)
df['Precipitación'] = df['Precipitación'].str.replace(r'[^\d.,]', '', regex=True).str.replace(',', '.').replace('', '0').astype(float)
```

```python
df.to_csv('Data.csv', index=False)
```

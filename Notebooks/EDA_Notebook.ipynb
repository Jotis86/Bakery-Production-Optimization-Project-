```python
#Data Visualitation
```

```python
# Function to save plots
def save_plot(fig, filename):
    fig.savefig(filename, bbox_inches='tight')
```

```python
# 1. Distribution of Sales by Hour
fig, ax = plt.subplots(figsize=(10, 6))
sns.countplot(data=df, x='Hora', ax=ax)
ax.set_title('Distribution of Sales by Hour')
ax.set_xlabel('Hour')
ax.set_ylabel('Number of Sales')
save_plot(fig, 'sales_by_hour.png')
```

```python
# 2. Total Sales by Day of the Week
fig, ax = plt.subplots(figsize=(10, 6))
sns.countplot(data=df, x='Dia_semana', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ax=ax)
ax.set_title('Total Sales by Day of the Week')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Number of Sales')
save_plot(fig, 'sales_by_day_of_week.png')
```

```python
# 3. Total Amount by Day of the Week
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=df, x='Dia_semana', y='Importe', estimator=sum, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ax=ax)
ax.set_title('Total Amount by Day of the Week')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Total Amount')
save_plot(fig, 'amount_by_day_of_week.png')
```

```python
# 4. Total Amount by Month
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=df, x='Mes', y='Importe', estimator=sum, ax=ax)
ax.set_title('Total Amount by Month')
ax.set_xlabel('Month')
ax.set_ylabel('Total Amount')
save_plot(fig, 'amount_by_month.png')
```

```python
# 5. Top 5 Most Sold Articles
top_5_articles = df['Articulo'].value_counts().nlargest(5).index
df_top_5 = df[df['Articulo'].isin(top_5_articles)]

fig, ax = plt.subplots(figsize=(10, 6))
sns.countplot(data=df_top_5, x='Articulo', ax=ax)
ax.set_title('Top 5 Most Sold Articles')
ax.set_xlabel('Article')
ax.set_ylabel('Number of Sales')
save_plot(fig, 'top_5_articles.png')
```

```python
# 6. Total Amount by Top 5 Articles
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=df_top_5, x='Articulo', y='Importe', estimator=sum, ax=ax)
ax.set_title('Total Amount by Top 5 Articles')
ax.set_xlabel('Article')
ax.set_ylabel('Total Amount')
save_plot(fig, 'amount_by_top_5_articles.png')
```

```python
# 7. Total Amount by Day of the Week and Holiday
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=df, x='Dia_semana', y='Importe', hue='Es_Festivo', estimator=sum, ax=ax)
ax.set_title('Total Amount by Day of the Week and Holiday')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Total Amount')
save_plot(fig, 'amount_by_day_and_holiday.png')
```

```python
# Filter out non-holiday data
df_holidays = df[df['Festivo'] != 'Día no festivo']

# 8. Sales by Type of Holiday
fig, ax = plt.subplots(figsize=(12, 8))
sns.countplot(data=df_holidays, y='Festivo', order=df_holidays['Festivo'].value_counts().index, ax=ax)
ax.set_title('Number of Sales by Type of Holiday')
ax.set_xlabel('Number of Sales')
ax.set_ylabel('Holiday')
plt.tight_layout()
plt.savefig('sales_by_holiday_type.png')
plt.show()

# 9. Total Amount by Type of Holiday
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=df_holidays, y='Festivo', x='Importe', estimator=sum, order=df_holidays['Festivo'].value_counts().index, ax=ax)
ax.set_title('Total Amount by Type of Holiday')
ax.set_xlabel('Total Amount')
ax.set_ylabel('Holiday')
plt.tight_layout()
plt.savefig('amount_by_holiday_type.png')
plt.show()

# 10. Sales by Hour on Holidays
fig, ax = plt.subplots(figsize=(12, 8))
sns.countplot(data=df_holidays, x='Hora', hue='Festivo', ax=ax)
ax.set_title('Sales by Hour on Holidays')
ax.set_xlabel('Hour')
ax.set_ylabel('Number of Sales')
plt.legend(title='Holiday', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('sales_by_hour_on_holidays.png')
plt.show()
```

```python
# 11. Total Sales Over Time
fig, ax = plt.subplots(figsize=(12, 8))
df.groupby('Fecha').size().plot(ax=ax)
ax.set_title('Total Sales Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Number of Sales')
plt.tight_layout()
plt.savefig('total_sales_over_time.png')
plt.show()
```

```python
# 12. Sales by Family Over Time
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(data=df, x='Fecha', y='Cantidad', hue='Familia', estimator=sum, ax=ax)
ax.set_title('Sales by Family Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Number of Sales')
plt.tight_layout()
plt.savefig('sales_by_family_over_time.png')
plt.show()
```

```python
# Filter the data to include only MERMA
df_merma = df[df['Tipo'] == 'MERMA']

# 13. Distribution of MERMA by Family
fig, ax = plt.subplots(figsize=(12, 8))
sns.countplot(data=df_merma, x='Familia', order=df_merma['Familia'].value_counts().index, ax=ax)
ax.set_title('Distribution of MERMA by Family')
ax.set_xlabel('Family')
ax.set_ylabel('Number of MERMA')
plt.tight_layout()
plt.savefig('merma_by_family.png')
plt.show()

# 14. Total Amount of MERMA by Day of the Week
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=df_merma, x='Dia_semana', y='Importe', estimator=sum, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ax=ax)
ax.set_title('Total Amount of MERMA by Day of the Week')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Total Amount')
plt.tight_layout()
plt.savefig('merma_by_day_of_week.png')
plt.show()

# 15. MERMA Over Time
fig, ax = plt.subplots(figsize=(12, 8))
df_merma.groupby('Fecha').size().plot(ax=ax)
ax.set_title('MERMA Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Number of MERMA')
plt.tight_layout()
plt.savefig('merma_over_time.png')
plt.show()
```

```python
# 16. Pie Chart: Proportion of MERMA by Family
plt.figure(figsize=(8, 8))
df_merma.groupby('Familia')['Cantidad'].sum().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis', len(df_merma['Familia'].unique())))
plt.title('Proportion of MERMA by Family', fontsize=16)
plt.ylabel('')
plt.tight_layout()
plt.savefig('merma_proportion_by_family.png')
plt.show()
```

```python
# 17. Line Chart: Trend of MERMA Over Time by Family and Year
plt.figure(figsize=(12, 8))
sns.lineplot(x='Año', y='Cantidad', hue='Familia', data=df_merma, palette='viridis', marker='o')
plt.title('Trend of MERMA Over Time by Family and Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Quantity', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Family')
plt.tight_layout()
plt.savefig('merma_trend_by_family_and_year.png')
plt.show()
```

```python
# 18. Bar Chart: Quantity of MERMA by Family and Year
plt.figure(figsize=(12, 8))
sns.barplot(x='Año', y='Cantidad', hue='Familia', data=df_merma, palette='viridis')
plt.title('Quantity of MERMA by Family and Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Quantity', fontsize=14)
plt.legend(title='Family')
plt.tight_layout()
plt.savefig('merma_quantity_by_family_and_year.png')
plt.show()
```

```python
# Define temperature ranges
def categorize_temperature(temp):
    if temp < 10:
        return 'Low'
    elif 10 <= temp < 20:
        return 'Medium'
    else:
        return 'High'

df['Temperature_Range'] = df['Temperatura'].apply(categorize_temperature)

# Categorize precipitation
df['Precipitation_Category'] = df['Precipitación'].apply(lambda x: 'Yes' if x > 0 else 'No')
```

```python
# 19. Box Plot: Sales Distribution by Temperature Range
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Temperature_Range', y='Importe', palette='viridis')
plt.title('Sales Distribution by Temperature Range')
plt.xlabel('Temperature Range')
plt.ylabel('Total Sales Amount')
plt.tight_layout()
plt.savefig('sales_distribution_by_temperature_range.png')
plt.show()
```

```python
# 20. Barplot: Sales Distribution by Precipitation Category
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Precipitation_Category', y='Importe', palette='viridis')
plt.title('Sales Distribution by Precipitation Category')
plt.xlabel('Precipitation Category')
plt.ylabel('Total Sales Amount')
plt.tight_layout()
plt.savefig('sales_distribution_by_precipitation_category.png')
plt.show()
```

```python
# 21. Barplot: Sales Distribution by Precipitation Category and Day of the Week
plt.figure(figsize=(12, 12))
sns.barplot(data=df, x='Precipitation_Category', y='Importe', hue='Dia_semana', palette='viridis')
plt.title('Sales Distribution by Precipitation Category and Day of the Week')
plt.xlabel('Precipitation Category')
plt.ylabel('Total Sales Amount')
plt.tight_layout()
plt.savefig('sales_distribution_by_precipitation_category_and_day.png')
plt.show()
```

```python
# 22. Barplot: Sales Distribution by Temperature Range and Day of the Week
plt.figure(figsize=(12, 12))
sns.barplot(data=df, x='Temperature_Range', y='Importe', hue='Dia_semana', palette='viridis')
plt.title('Sales Distribution by Temperature Range and Day of the Week')
plt.xlabel('Temperature Range')
plt.ylabel('Total Sales Amount')
plt.tight_layout()
plt.savefig('sales_distribution_by_temperature_range_and_day.png')
plt.show()
```

```python
# 23. Heatmap: Correlation Between Variables
plt.figure(figsize=(10, 6))
corr = df[['Temperatura', 'Precipitación', 'Importe']].corr()
sns.heatmap(corr, annot=True, cmap='viridis')
plt.title('Correlation Between Variables')
plt.tight_layout()
plt.savefig('correlation_between_variables.png')
plt.show()
```
